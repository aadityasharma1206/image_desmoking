```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

```python
for i in range(1, 5501):
    main_img = cv2.imread(f"Dehazing Datasets/ILSVRC2017_DET_test_new/ILSVRC/Data/DET/test/ILSVRC2017_test_{i:08d}.jpeg")
    main_img = cv2.cvtColor(main_img, cv2.COLOR_BGR2RGB)
    smoke_img = cv2.imread("Dehazing Datasets/smoke1.jpeg")
    smoke_img = cv2.cvtColor(smoke_img, cv2.COLOR_BGR2RGB)
    smoke_img = cv2.resize(smoke_img, (main_img.shape[1], main_img.shape[0]))
    x = random()
    if(x >= 0.7):
        x -= 0.2
    elif(x < 0.5):
        x = 0.5
    result = cv2.addWeighted(src1=main_img, alpha=0.9, src2=smoke_img, beta = x, gamma=0)
    result = cv2.cvtColor(result, cv2.COLOR_RGB2BGR)
    cv2.imwrite(f"Dehazing Datasets/smoked/img_{i:08d}.jpeg", result)
```

```python
# example
img = cv2.imread("Dehazing Datasets/ILSVRC2017_DET_test_new/ILSVRC/Data/DET/test/ILSVRC2017_test_00000263.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
